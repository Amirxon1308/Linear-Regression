## Linear Regression

Regressiya o'zgaruvchilar o'rtasidagi bog'liqlikni aniqlash uchun ishlatiladi. 
Machine Learningda, statistikada, data scienceda bog'liqlik kelajakdagi natijalarni bashorat qilish uchun ishlatiladi. 


Linear regression data ustunlari o'rtasidagi bog'liqlarga ko'ra ularda tog'ri chiziq chizadi va bu chiziq kelajada yangi natijalarni bashorati uchun ishlatiladi. 


Python dasturlash tilida bog'liqlikni aniqlab beradigan va shu bog'liqlik bo'yicha linear regressiya chiziqlarini chizib beradigan maxsus methodlar mavjud.


Masala:
    x o'qi bilan mashinaning yoshi berilgan, y-o'qi esa mashinaning tezligini belgilaydi. 
    Linear regression yordamida mavjud bog'liqlikni tekshirish lozim.

```python
import matplotlib.pyplot as plt

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]   #mashina foydalanilgan yillar
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]  #mashina tezligi

plt.scatter(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy import stats   # linear regressionni ifodalash uchun

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]   #x va y qiymatlari uchun arraylar

slope, intercept, r, p, std_err = stats.linregress(x, y)

#slope va intercept yordamida yangi qiymatlarni qaytarish uchun, 
#bu funksiya x o'qiga mos y o'qini joylashtirib beradi 


def myfunc(x):
    return slope * x + intercept


mymodel = list(map(myfunc, x)) #x arrayining har bitta funsiyasi bo'yicha hisoblashni
#amalga oshirish uchun. Natijada y o'qi uchun yangi arraylarni topish mumkin. 

plt.xlabel("Mashina yoshi")
plt.ylabel("Tezligi")

plt.scatter(x, y) #original scatter plot


plt.plot(x, mymodel) # linear regression chizig'ini chizish

plt.show()
```

# R for Relationship 
# Bog'liqlikni tashvirlash: R

x-axes va y-axes o'rtasidagi bog'laiqliqlikni aniqlash judayam muhim, agar bog'liqlik bo'lmasa yangi qiymatlarni bashorat qilib bo'lmaydi. 
Bog'liqlikni belgilash uchun r dan foydalaniladi. 

r ning qiymatlari -1 dan 1 gacha, agar r=0 bo'lsa hech qanday bog'liqlik mavjud emas.
Agar bog'liqlik 1 yoki -1 bo'lsa 100% bog'liqlik mavjud. 

```python
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

print(r)
```

Bog'liqlik judayam yaxshi emas (76%) lekin bu bog'liqlik yangi qiymatlarni bashorat qilish uchun ishlatilishi mumkin.

## Predict Future Values - Kelajakdagi qiymatlarni bashorat qilish

Endi yuqoridagi bog'liqlik asosida turli yildagi mashinalar uchun bo'lishi mumkin bo'lgan tezlikni bashorat qilish mumkin.


```python
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
    return slope * x + intercept

speed = myfunc(10)

print(speed)
```

```python
speed = myfunc(15)

print(speed)
```
